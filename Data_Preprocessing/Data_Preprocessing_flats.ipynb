```python
# lets create the function who scale down price


def treat_price(x):
  if type(x) == float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)


```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df['price']
```

```python
# still null values are presents so we can treat later on
df['price'].isnull().sum()
```
