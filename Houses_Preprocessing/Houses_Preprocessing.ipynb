#Price

```python
df['price'].value_counts()
```

```python
# we noticed that Price on Request are in the value_counts
df = df[df['price'] != 'Price on Request']
```

```python
df.shape
```

#### Observation:- we noticed we left 1021 rows

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df['price']
```
